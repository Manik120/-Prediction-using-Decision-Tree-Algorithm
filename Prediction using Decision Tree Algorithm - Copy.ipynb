```python
#Importing the requried library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree ,datasets
```

```python
#Getting the dataset
iris_data=pd.read_csv("D:\\INTERNSHIP\\fDecision tree\\Iris.csv")
iris_data.head()
```

```python
iris_data=datasets.load_iris()
iris_df=pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
y=iris_data.target
print("The first five entries of data\n", iris_df.head(5))
print("The clssvalues\n",y)
```

```python
#Define the data
x=iris_df.to_numpy()
#Defining the decision tree Classifier
dt_model=tree.DecisionTreeClassifier(random_state=0)
dt_model.fit(x, y)

fig=plt.figure(figsize=(27,20), facecolor="k")
dtree=tree.plot_tree(dt_model,
                    feature_names=iris_data.feature_names,
                    class_names=iris_data.target_names,
                    filled=True,
                    rounded=True,
                    fontsize=20)

#Customising the arrow 
for object in dtree:
    arrow=object.arrow_patch
    if arrow is not None:
        arrow.set_edgecolor("yellow")
        arrow.set_linewidth(5)
```
